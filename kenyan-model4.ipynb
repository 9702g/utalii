Handling Missing Values:
1.data['travel_with'] = data['travel_with'].fillna('Alone'):
•Fills missing values in the travel_with column with the string "Alone". This assumes that "Alone" is a reasonable default value for this column. 
2.data['total_female'] = data['total_female'].fillna(1.0):
•Fills missing values in the total_female column with the number 1.0. This assumes there's at least one female present when the value is missing.This is accurate according to the data . 
3.data['total_male'] = data['total_male'].fillna(1.0):
•Similar to above, fills missing values in the total_male column with 1.0. 
4.data['most_impressing'] = data['most_impressing'].fillna('No comments'):
•Replaces missing values in the most_impressing column with the string "No comments". 
5.data['age_group'] = data['age_group'].replace('24-Jan', '1-24'):
•Replaces occurrences of "24-Jan" in the age_group column with "1-24". 


```python
# Clearing errors and missing values from data set
# For the travel_with and most_impressing columns, fill in with a new category, None
data['travel_with'] = data['travel_with'].replace(np.nan, 'Alone')
#replacing the nan values in the total_female column of the training data with 1.0

data['total_female'] = data['total_female'].replace(np.nan, 1.0)
#replacing the nan values in the total_male column of the training data with 1.0

data['total_male'] = data['total_male'].replace(np.nan, 1.0)

#replacing the nan values in the most_impressing column of the training data with No comments

data['most_impressing'] = data['most_impressing'].replace(np.nan, 'No comments')
#replacing 24-jan to 1-24 to correlate with 

data['age_group'] = data['age_group'].replace('24-Jan', '1-24')
# %convert float dtypes to int[total_female,total_male,night_mainland,night_zanzibar]
data["total_female"] = data['total_female'].astype('int')
data["total_male"] = data['total_male'].astype('int')
data["nights_stayed"] = data['nights_stayed'].astype('int')
# %Let's generate new features from some columns which makes some sense
data["total_people"] = data["total_female"] + data["total_male"]


#chech for missing values
data.isnull().sum()
# %then it's time to encode objects into numeric
```

Adjusted R2 Score: It adjusts the R2 score to penalize the addition of unnecessary independent variables in the model.

Mean Absolute Error (MAE): It calculates the absolute error between each predicted value and its corresponding actual value, then takes the mean of these absolute errors.

Interpretation:
